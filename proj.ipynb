A partir desse ponto, tomamos a decisão de unificar as bases a partir de suas PKs para facilitar a análise, já que a base contém apenas 96k de linhas

```python
df_m = pd.merge(df, df1, on='customer_id', how='inner')
```

Verificando se o merge deu certo antes de fazer os outros...

```python
df_m.info()
```

Deu certo! Seguindo com os outros merges! Faltam a 2, 3, 4, 5 e 6. Começaremos com a 3, pois a PK da 2 é o seller_id que ainda não temos acesso.

```python
df_m = pd.merge(df_m, df3, on='order_id', how='inner')
```

```python
df_m = pd.merge(df_m, df4, on='order_id', how='inner')
```

```python
df_m = pd.merge(df_m, df6, on='order_id', how='inner')
```

```python
df_m = pd.merge(df_m, df2, on='seller_id', how='inner')
```

```python
df_m = pd.merge(df_m, df5, on='product_id', how='inner')
```
